```python
#Eliminnando variveis que no sero utilizadas
df1 = df.drop(columns =['RowNumber', 'CustomerId', 'Surname'])
df1
```

```python
# separando variveis
Y = df1['Exited']
X = df1
X = df1.drop(columns = ['Exited'])
X
```

```python
#separando vairveis categricas e numricas
#variveis continuas
x_cont =['CreditScore', 'Balance', 'Age', 'EstimatedSalary', 'Tenure']
```

```python
#variveis categricas
x_cat = list(set(X)- set (x_cont))
x_cat
x_dumm = X[x_cat]
x_dumm
```

```python
#trnasformando as variveirs stg em numricas
#substituindo varivel sexo para 0 e 1
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])
X.head(10)
```

```python
x_final =pd.get_dummies (data = X, columns=['Geography', 'NumOfProducts'])
x_final
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_final[x_cont] =scaler.fit_transform(x_final[x_cont])
x_final[x_cont]
```

Modelagem

```python
from sklearn.model_selection import train_test_split
split_test_size = 0.25
X_treino, X_teste, Y_treino, Y_teste = train_test_split(x_final, y, test_size = split_test_size, random_state = 1)
X_treino
```

```python
#modelo KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_treino, Y_treino)
```

```python
reultado_knn = knn.predict(X_teste)
print(pd.crosstab(Y_teste,resultado_knn))
```

```python
from sklearn import metrics
```

```python
nb_predict_train = knn.predict(X_treino)
```

```python
#verificando a exatido do modelo
print(f'Exatido (Accuracy) do modelo: {metrics.accuracy_score(Y_treino, nb_predict_train)*100:.2f} %')
```

```python
nb_predict_teste = knn.predict(X_teste)
```

```python
#verificando a exatido do modelo
print(f'Exatido (Accuracy) do modelo: {metrics.accuracy_score(Y_teste, nb_predict_teste)*100:.2f} %')
```

```python
# Criando uma Confusion Matrix
print("Confusion Matrix")

print("{0}".format(metrics.confusion_matrix(Y_teste, nb_predict_teste, labels = [1, 0])))
print("")

print("Classification Report")
print(metrics.classification_report(Y_teste, nb_predict_teste, labels = [1, 0]))
```

```python
error =[]
for i in range(1,40):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_treino, Y_treino)
    pred_i= knn.predict(X_teste)
    error.append(np.mean(pred_i !=Y_teste))
```
